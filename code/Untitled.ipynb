```python
import fsys
import cv2
fsys.cd('/Users/colin/Dropbox/__Atlas__/data/07119/histology')
img1 = cv2.imread('crsSlice003.tiff')
img2 = cv2.imread('crsSlice004.tiff')
```

```python
import numpy as np
import imutils
from scipy.stats.stats import pearsonr 
p_xytheta = np.zeros([50,50,50])
ix = 0
for x in np.linspace(-10,10):
    iy=0
    for y in np.linspace(-10,10):
        itheta = 0
        print([x,y,theta])
        for theta in np.linspace(-10,10):
            newimg2 = imutils.rotate(img2,theta)
            newimg2 = imutils.translate(newimg2,x,y)
            p_xytheta[ix,iy,itheta],p = pearsonr(img1.flatten(),newimg2.flatten())
            itheta+=1
            
        iy+=1
    ix+=1
print('done')
            
```

```python
p_xytheta = p_xytheta/np.sum(p_xytheta.flatten())
```
