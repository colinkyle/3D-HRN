```python
import numpy as np
import pylab as pl
from scipy import stats

# Size of dataset
size = 5000
# Predictor variable
X1 = 10*np.random.randn(size)
px1 = stats.laplace.pdf(X1,loc = -1, scale = 3)

X2 = 10*np.random.randn(size)
scale2 = 20
px2 = stats.triang.pdf(X2,.5,loc = 6-(scale2/2), scale = scale2)

X3 = 10*np.random.randn(size)
scale3 = 15
px3 = stats.triang.pdf(X3,.5,loc = -3-(scale3/2), scale = scale3)
# Simulate outcome variable
Y = px1*px2*px3 #+ 0.00001*np.random.randn(size)*sigma

%pylab inline
fig, axes = subplots(1, 3, sharex=True, figsize=(10,4))
axes[0].scatter(X1, px1)
axes[1].scatter(X2, px2)
axes[2].scatter(X3, px3)
axes[0].set_ylabel('Y'); axes[0].set_xlabel('X1'); axes[1].set_xlabel('X2');
```

```python
import pymc3 as pm
from scipy import optimize
import math
from theano import tensor as T
import theano

PI = np.array(np.pi, dtype=theano.config.floatX)
epsilon = np.array(1e-32, dtype=theano.config.floatX)
X1a = np.array(X1, dtype=theano.config.floatX)
X2a = np.array(X2, dtype=theano.config.floatX)
X3a = np.array(X3, dtype=theano.config.floatX)

def gpdf(sample, location, scale):

    SQRT_2_PI = np.sqrt(2 * PI)
    SQRT_2_PI = T.cast(SQRT_2_PI, theano.config.floatX)

    divisor = 2 * scale ** 2 # + epsilon,
#     if isinstance(location, T.TensorVariable) and location.ndim == 0:
#         location = location.dimshuffle('x')

    exp_arg = -((sample - location) ** 2) / divisor
    z = 1. / (SQRT_2_PI * scale + epsilon)

    return T.exp(exp_arg) * z

def tpdf(sample, location, scale):
#     if isinstance(location, T.TensorVariable) and location.ndim == 0:
#         location = location.dimshuffle('x')
    m = -1/(scale ** 2)
    b = 1/scale
    return m*T.abs_(sample - location) + b

def lpdf(sample, location, scale):

    divisor = 2 * scale# + epsilon,
#     if isinstance(location, T.TensorVariable) and location.ndim == 0:
#         location = location.dimshuffle('x')

    exp_arg = -T.abs_(sample - location) / divisor
    z = 1. / (2 * scale + epsilon)

    return T.exp(exp_arg) * z

basic_model = pm.Model()

with basic_model:

    # Priors for unknown model parameters
    m1 = pm.Normal('m1',mu=0,sd=10)
    m2 = pm.Normal('m2',mu=0,sd=10)
    m3 = pm.Normal('m3',mu=0,sd=10)
    
    s1 = pm.HalfNormal('s1', sd=1)
    s2 = pm.HalfNormal('s2', sd=1)
    s3 = pm.HalfNormal('s3', sd=1)
    
    #alpha = pm.Laplace('x1', mu=m1, b=s1)
    alpha = lpdf(X1a,m1,s1)
    beta = tpdf(X2a,m2,s2)
    zeta = tpdf(X3a,m3,s3)
    #alpha = stats.laplace.pdf(X1,loc = m1, scale = s1)
    #beta = pm.Normal('x2', mu=m2, sd=s2)
    #beta = stats.normal.pdf(X2, loc=m2, scale=s2)
    #zeta = stats.normal.pdf(X3, loc=m3, scale=s3)
    
    sigma = pm.HalfNormal('sigma', sd=1)

    # Expected value of outcome
    mu = alpha*beta*zeta

    # Likelihood (sampling distribution) of observations
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=sigma, observed=Y)
    trace = pm.sample(3000,njobs=4)
```

```python
import cv2
import numpy as np
# Read the images to be aligned
im1 =  cv2.imread("../data/30890/histology/30890 B1S2 220N_01.tiff");
im2 =  cv2.imread("../data/30890/histology/30890 B1S2 222N_01.tiff");
 
# Convert images to grayscale
im1_gray = cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY)
im2_gray = cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)
 
# Find size of image1
sz = im1.shape
 
# Define the motion model
warp_mode = cv2.MOTION_AFFINE
 
# Define 2x3 or 3x3 matrices and initialize the matrix to identity
if warp_mode == cv2.MOTION_HOMOGRAPHY :
    warp_matrix = np.eye(3, 3, dtype=np.float32)
else :
    warp_matrix = np.eye(2, 3, dtype=np.float32)
 
# Specify the number of iterations.
number_of_iterations = 5000;
 
# Specify the threshold of the increment
# in the correlation coefficient between two iterations
termination_eps = 1e-2;
 
# Define termination criteria
criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, number_of_iterations,  termination_eps)
 
# Run the ECC algorithm. The results are stored in warp_matrix.
(cc, warp_matrix) = cv2.findTransformECC (im1_gray,im2_gray,warp_matrix, warp_mode, criteria)
 
if warp_mode == cv2.MOTION_HOMOGRAPHY :
    # Use warpPerspective for Homography 
    im2_aligned = cv2.warpPerspective (im2, warp_matrix, (sz[1],sz[0]), flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP)
else :
    # Use warpAffine for Translation, Euclidean and Affine
    im2_aligned = cv2.warpAffine(im2, warp_matrix, (sz[1],sz[0]), flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP);
 
# Show final results
#cv2.imshow("Aligned Image 2", im2_aligned)
#cv2.waitKey(0)
```
